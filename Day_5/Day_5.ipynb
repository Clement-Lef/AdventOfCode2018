# Part 1

```python
filename = 'input.txt'
with open(filename) as f:
    data = f.readlines()
f.close()
for n,line in enumerate(data):
    data[n] = line.rstrip()
```

```python
import re
```

```python
# Test
test_string = 'dabAcCaCBAcCcaDA'
reaction = re.compile(chr(ord("c")+0) + chr(ord("C")+0))
re.subn(reaction,'',test_string)
```

```python
new_string = data[0]
num_of_subs = 1
while num_of_subs > 0:
    num_of_subs = 0
    for i in range(26):
        (new_string, num_of_sub) = re.subn(re.compile(chr(ord("a")+i) + chr(ord("A")+i) + '|' +
                                                     chr(ord("A")+i) + chr(ord("a")+i)), '', new_string)
        num_of_subs += num_of_sub
    
print(len(new_string))
```

```python
reduced_string = new_string

length_new_string_per_unit = []

for i in range(26):
    new_string = reduced_string
    (new_string, num_of_sub) = re.subn(re.compile(chr(ord("a")+i)+ '|' + chr(ord("A")+i)),'',new_string)
    num_of_subs = 1
    while num_of_subs > 0:
        num_of_subs = 0
        for j in range(26):
            (new_string, num_of_sub) = re.subn(re.compile(chr(ord("a")+j) + chr(ord("A")+j) + '|' +
                                                     chr(ord("A")+j) + chr(ord("a")+j)), '', new_string)
            num_of_subs += num_of_sub
    length_new_string_per_unit.append(len(new_string))
```
